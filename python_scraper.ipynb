```python
from bs4 import BeautifulSoup as bs
import pandas as pd
pd.set_option('display.max_colwidth', 500)
import requests
import re

#request URL
page = requests.get("https://www.nhl.com/standings/2021/conference")
page
# <Response [200]>

parse = bs(page.content)
parse

# print all <span> tags with team names
for team in parse.find_all('span', class_=re.compile('top-nav__secondary-section__menu__item__link__team-name top-nav__secondary-section__menu__item__link--team-*')):
    print(team)
    


```

```python
# now we can start adding the pulled data to a pandas dataframe
teamNames = []
#get_text() method to get just the team names:
for team in parse.find_all('span', class_=re.compile('top-nav__secondary-section__menu__item__link__team-name top-nav__secondary-section__menu__item__link--team-*')):
    print(team.get_text())
    
    teamNames.append(team.get_text())
    
```

```python
print(teamNames)

#convert list to pandas dataframe
df = pd.DataFrame(teamNames)
df.columns = ['Team']
df
```
